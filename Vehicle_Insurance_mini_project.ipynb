###**PREMIUM ANALYSIS :**

```python
plt.figure(figsize=(10,6))
sns.histplot(data['Annual_Premium'], kde=True, bins=30, color='blue')
plt.title("Annual Premium Distribution", fontsize=15, fontweight="bold")
plt.xlabel("Premium Amount",fontsize=15, fontweight="bold")
plt.ylabel("Count",fontsize=15,fontweight="bold")
plt.show()
```

**Insights:**

There is no significant difference in annual premium amounts between customers who responded and those who didn’t, though both groups contain many high-value outliers.


FIXING THE OUTLIERS:

```python
Q1 = data['Annual_Premium'].quantile(0.25)
Q3 = data['Annual_Premium'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

data['Annual_Premium'] = data['Annual_Premium'].clip(lower=lower_bound, upper=upper_bound)
```

```python
# Filter the dataset to remove outliers
data = data[(data['Annual_Premium'] >= lower_bound) & (data['Annual_Premium'] <= upper_bound)]
```

CHECKING THE OUTLIER AGAIN:

```python
plt.figure(figsize=(10,6))
sns.boxplot(x='Response',y='Annual_Premium',data=data)
plt.title("Annual Premium vs Insurance Claim", fontsize=15, fontweight="bold")
plt.xlabel("Response",fontsize=15, fontweight="bold")
plt.ylabel("Annual_Premium",fontsize=15,fontweight="bold")
plt.show()
```

**Insights:**

After handling outliers, the annual premium distribution appears more balanced across both response groups, with no major difference in median premium values between customers who claimed and those who didn’t.
